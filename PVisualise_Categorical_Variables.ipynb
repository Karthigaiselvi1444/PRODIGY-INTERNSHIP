```python
#Importing the required libraries:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Df = pd.read_csv(r"D:\PRODIGY INTERNSHIP\API_SP.POP.TOTL_DS2_en_csv_v2_84031\API_SP.POP.TOTL_DS2_en_csv_v2_84031.csv")
Df.head()
```

```python
#Dropping the columns that are not needed:
Df_Final=Df.drop(['Country Code','Indicator Name','Indicator Code'],axis=1)
print(Df_Final.info())
```

```python
#Picking up one country for visualizing the bar and histogram:
#print(Df_Final['Country Name'].unique())
Chosen_country='United Arab Emirates'
Country_Data=Df_Final[Df_Final['Country Name']==Chosen_country]

#Formatting:
Years=Country_Data.columns[2:].astype('int')
Population=Country_Data.iloc[:,2:].values.flatten()

#Plotting:
plt.figure(figsize=(8,8))
plt.bar(Years,Population)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title(f'Population of {Chosen_country} for the respective years')
plt.show()
```

```python
#Plotting histogram:
plt.hist(Population,edgecolor='black')
plt.show()
```
