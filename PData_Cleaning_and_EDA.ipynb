```python
#Importing the required libraries:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Df=pd.read_csv(r"D:\PRODIGY INTERNSHIP\Titanic_Dataset.csv")
print(Df.head())
```

```python
#Performing the Data-Cleaning steps:
#Statisical analysis:
print(Df.describe())
```

```python
#Describing the dataset:
print(Df.info())
```

```python
#Checking whether there are null values:
print('Column wise null values:',Df.isnull().sum())
print('Total null values:',Df.isnull().sum().sum())
```

```python
#Filling the null values with the mean value of each column:
Df['Age'].fillna(Df['Age'].mean,inplace=True)
Df['Cabin'].fillna(Df['Cabin'].mean,inplace=True)
Df['Embarked'].fillna(Df['Embarked'].mean,inplace=True)
print('Column wise null values:',Df.isnull().sum())
print(Df.shape)
```

```python
#Checking whether there are any duplicate values:
print(Df.duplicated().sum())
```

```python
# Convert 'Age' column to numeric, coerce errors to NaN
Df['Age'] = pd.to_numeric(Df['Age'], errors='coerce')

# Drop NaN values
Df = Df.dropna(subset=['Age'])

# Plotting
plt.figure(figsize=(8,8))
sns.histplot(Df['Age'])
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
#Count-Plot:
plt.figure(figsize=(8,8))
sns.countplot(data=Df,x="Sex",hue="Survived")
plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Survived",loc="upper right")
plt.show()
```

```python
#Scatter-Plot
plt.figure(figsize=(8,8))
sns.scatterplot(data=Df,x=Df["Age"],y=Df["Fare"],hue=Df["Survived"])
plt.title("Scatter Plot of Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend(title="Survived")
plt.show()
```
